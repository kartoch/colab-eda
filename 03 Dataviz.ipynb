visualize a linear relationship as determined through regression: [regplot](https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot) and [lmplot](https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot)

Using the 'iris' dataset, examples are extracted from [The python graph library](https://python-graph-gallery.com)

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = sns.load_dataset('iris')
```

#### Simple scatterplot

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
```

#### Marginal plot

```python

sns.jointplot(x="sepal_length", y="sepal_width", kind='scatter', data=df)
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", kind='hex', data=df)
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", kind='kde', data=df)
```

#### Change aspects

```python
sns.lmplot( x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False, markers=["o", "x", "1"])
plt.legend(loc='lower right')
```

#### Bivariate regression

```python
sns.regplot( x="sepal_length", y="sepal_width", data=df)
```

#### Multilinear regression

```python
sns.lmplot(x="sepal_length", y="sepal_width", hue="species", data=df, truncate=True)
```

```python
sns.pairplot(data=df, hue="species", height=2.5)
```

#### Distribution

```python
sns.boxplot(x='species', y='sepal_length', data=df)
```

```python
sns.stripplot(x='species', y='sepal_length', data=df)
```

```python
sns.swarmplot(x='species', y='sepal_length', data=df)
```

```python
ax = sns.boxplot(x='species', y='sepal_length', data=df)
sns.swarmplot(x='species', y='sepal_length', data=df, color="grey",ax=ax)
```

```python
sns.violinplot(x="species", y="sepal_length", data=df)
```

```python
sns.boxenplot(x="species", y="sepal_length", data=df)
```

```python
sns.catplot(data=df, orient="v", kind="box");
```

## Interactive vizualisation

* **bokeh** allows the easy creation of interactive data visualizations.

### Bokeh

* create interactive vizualization in browsers without javascript
* can be used with dynamic and streaming data
* low and high-level API for easy rendering

```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()
```

```python
colormap = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
colors = [colormap[x] for x in df['species']]

p = figure(title = "Iris Morphology")
p.xaxis.axis_label = 'Petal Length'
p.yaxis.axis_label = 'Petal Width'

p.circle(df["petal_length"], df["petal_width"], color=colors, fill_alpha=0.2, size=10)

show(p)
```
